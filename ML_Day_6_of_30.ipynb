### **Feature Engineering**

**Feature Engineering:**  
Process of using domain knowledge to extract features from raw data. These features can be used to improve the performance of ML Algorithms.  


image.png

We can't give data directly to ML model. Instead we should give data to prepare then give to ML model.  
Process of converting RAW data into a form of data which ML models can consume is called as Feature Engineering.

```
Feature Engineering:
|
|-----> Feature Transformation
|
|-----> Feature Construction
|
|-----> Feature Selection
|
|-----> Feature Extraction
```



**Feature Transformation:**  
Construction of features/columns into another form such that it gets easy for us to further process so that it will perform better while training.  
  
**Feature Construction:**  
Sometimes the features we want are not available in our datasets, thus we have to create them manually. So that those features can give us better results.  
  
**Feature Selection:**  
We are not giving whole features from dataset to applied ML algorithms, instead we are giving only some selected features to it.  

**Feature Extraction:**  
Extracting new features from the given features programatically by using some sort of algorithms.

### **Standard Scaler**

````python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# fit the scaler to the train set, it will learn the parameters
scaler.fit(X_train)

# transform train and test sets
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
````

**scaler = StandardScaler()**: It creates an instance of the StandardScaler class, which is used for standardizing the features by removing the mean and scaling to unit variance. (We stored StandardScaler in  object scaler)

**scaler.fit(X_train)**: It fits the scaler to the training set (X_train), which calculates and stores the mean and standard deviation of each feature in the training set. This step is necessary to learn the scaling parameters.

**X_train_scaled = scaler.transform(X_train)**: It applies the learned scaling parameters to the training set (X_train) and transforms the features by subtracting the mean and dividing by the standard deviation. This standardizes the values to have zero mean and unit variance.  
z = (x - u) / s

**X_test_scaled = scaler.transform(X_test)**: It applies the same scaling parameters learned from the training set to the test set (X_test). This ensures that the test set is transformed using the same scaling as the training set, maintaining consistency in the feature scaling.  
z = (x - u) / s

The **accuracy_score** is a function provided by the scikit-learn library in Python. It is used to measure the accuracy or correctness of a classification model's predictions.

In machine learning, classification is the task of assigning labels or categories to data based on its features. The accuracy score helps evaluate how well the model performs in predicting these labels correctly.

The accuracy_score function compares the predicted labels with the true labels of a dataset and calculates the percentage of correct predictions. It takes two arguments: the true labels (y_true) and the predicted labels (y_pred).

The accuracy score is calculated as the ratio of the number of correct predictions to the total number of predictions:

Accuracy = (Number of Correct Predictions) / (Total Number of Predictions)

A high accuracy score indicates that the model's predictions are close to the true labels, implying that the model is performing well. However, accuracy alone may not provide a complete picture of model performance, especially when dealing with imbalanced datasets or certain types of classification problems.

It is important to consider other evaluation metrics and domain-specific requirements to get a comprehensive understanding of model performance. Some additional evaluation metrics include precision, recall, F1-score, and area under the ROC curve (AUC-ROC).

In summary, the accuracy score is a useful metric to assess the overall performance of a classification model and to compare different models. It helps in evaluating the model's ability to correctly predict the labels and provides a quick measure of its accuracy.

### **Mean Normalization:**

Range: -1 to 1  
Formula: x_i' = (x_i - x_mean) / (x_max - x_mean)  

If:  
value we get > mean, then its +ve  
value we get < mean, then its -ve

This is used in algorithms where we need centered data.

### **Max - Absolute Scaling:**

Formula: x_i' = (x_i) / (|x_max|)  
This is generally used for Sparse data.  
Sparse data is the data where there are more 0s present.  
In Sklearn it is available as 'MaxAbsScaler'.

###**Robust Scaling:**

Formula: x_i' = (x_i - x_median) / IQR {75th oercentile value - 25th percentile value}.  
IQR --> Inter Quartile Range. IQR = (Q3 - Q1), where Q3 is the upper quartile and Q1 is the lower quartile.   
In Sklearn it is available as 'Robustscaler'.  
We shold try this scaling if outliers are preent in our dataset.
