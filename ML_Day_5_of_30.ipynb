```python
# Finding percentage of Survived Passengers in pparticular class

Titanic_df.groupby('Pclass').mean()['Survived']*100
```

**groupby('Pclass')**: This groups the rows of the DataFrame based on the unique values in the 'Pclass' column. It creates separate groups for each passenger class.

**mean()**: This calculates the mean value for each column in the grouped DataFrame. Since the 'Survived' column contains binary values (0 or 1), taking the mean will give the proportion of passengers who survived in each passenger class.

**['Survived']**: This selects only the 'Survived' column from the grouped and mean-calculated DataFrame.

***100**: This multiplies the mean values (proportions) by 100 to convert them into percentages.

```python
# PLOT IT

(Titanic_df.groupby('Pclass').mean()['Survived']*100).plot(kind = 'bar')
```

```python
# How many percentage of Males were survived

Titanic_df.groupby('Sex').mean()['Survived']*100
```

```python
# What percent of passengers from particular station were survived

Titanic_df.groupby('Embarked').mean()['Survived']*100
```

### **Line Plot (Numerical - Numerical)**

Line Plot is used when we are comparing data which is related to time (It can be either date, time, month, year, etc)   
Joining all points of scatterplot results us to give lineplot in linear way.  
This we can use in stock prices, etc time related output for something...

```python
# For example taking an example dataset for Passengers travelling in flight from USA

Flights_df = pd.read_csv('/content/Air_Passengers_flight.csv')
```

```python
Flights_df
```

```python
# Split the 'Month' column into 'Year' and 'Month' columns
#Flights_df = Flights_df[['Year', 'Month']] = df['Month'].str.split('-', 1, expand=True)
```

```python
# Load flights dataset

Flights_df
```

```python
# Lets convert the dataset into 3 different columns 'Year', 'Month', 'Passengers'

# So firstly we split our month column which contains both year and month, where month is seperated by '-'

df = Flights_df[["month", "year"]] = Flights_df["Month"].str.split("-", expand = True)
```

```python
df
```

```python
# So as we created two columns for Year and Month, now we will add one more columnn to it where we will bw storing data about no. of Passengers

df['Passengers'] = pd.Series(Flights_df['#Passengers'])
```

```python
df
```

```python
# Our dataset is having column names '0', '1' and Passengers so we will update new column names for it as 'Year', 'Month' and 'Passengers'

df.columns = ['Year', 'Month', 'Passengers']
```

```python
df
```

```python
# Now we will assign month names instead of month numbers in Month coulumn.

# Define a dictionary mapping month numbers to month names

month_names = {
    '01' : 'January',
    '02' : 'February',
    '03' : 'March',
    '04' : 'April',
    '05' : 'May',
    '06' : 'June',
    '07' : 'July',
    '08' : 'August',
    '09' : 'September',
    '10' : 'October',
    '11' : 'November',
    '12' : 'December'
}

# Rename month numbers to month names in the 'Month' column
df['Month'] = df['Month'].replace(month_names)
```

```python
df
```

```python
# Let's find total no. of Passengers for particular year and particular month.

df.pivot_table(values = 'Passengers', index = 'Month', columns = 'Year')
```

```python
# Now let us analyze the data using heatmap

sns.heatmap(df.pivot_table(values = 'Passengers', index = 'Month', columns = 'Year'))
```

By analyzing the data we can see that yearly the no. of passengers are more in month of July and August.   
So, as a year is increasing the passengers are also increasing and mainly majority of passengers are travelling in month of July and August, then June, then September, then May, then April, and so on....

```python
# For more understanding of analyzation of data we will create clustermap.

sns.clustermap(df.pivot_table(values = 'Passengers', index = 'Month', columns = 'Year'))
```

As we can see in year 1959 & 1960 there were more no of Passengers, then year 1957 & 1958, then year 1955 and 1956, then 1953 & 1954, then 1951 & 1952 and then 1949 & 1950.  
If we see month wise in the graph we can more clearly analyze which month the more Passengers use to travel, that is July and August. And further more analyzations can be done.

```python
sns.lineplot(x = df['Year'], y = df['Passengers'])
```

```python
new_df = df.groupby('Year').sum().reset_index()
```

**df.groupby('Year')**: This groups the DataFrame df by the unique values in the 'Year' column. It creates a GroupBy object that can be used for aggregating data based on the groups.

**sum()**: This calculates the sum of all numeric columns within each group. It applies the sum function to each group and returns a new DataFrame with the summed values.

**reset_index()**: This resets the index of the resulting DataFrame. By default, when performing aggregation with groupby and applying a function like sum, the resulting DataFrame will have the grouped column(s) as the index. Calling reset_index() reverts the index back to the default integer index and adds the grouped column(s) as regular columns in the DataFrame.

new_df = This assigns the resulting DataFrame after grouping, summing, and resetting the index to the variable new_df.

```python
new_df
```

```python
# now we create linear graph for Passengers for particular Year

sns.lineplot(x = new_df['Year'], y = new_df['Passengers'])
```
